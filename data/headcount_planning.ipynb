```python
df = pd.read_csv("/Users/frdilla/Desktop/Strategic Headcount Planning/headcount_planning.csv", parse_dates=["start_date"])

rows = []
for _, row in df.iterrows():
    for m in range(int(row["tenure_months"])):
        rows.append({
            "employee_name": row["employee_name"],
            "payment_month": row["start_date"] + pd.DateOffset(months=m),
            "monthly_salary": row["monthly_salary"],
            "monthly_bonus": row["bonus"] / row["tenure_months"] if row["bonus_frequency"] == "Once" else row["bonus"],
        })

df_monthly = pd.DataFrame(rows)
df_monthly.to_csv("monthly_compensation_expanded.csv", index=False)
```
